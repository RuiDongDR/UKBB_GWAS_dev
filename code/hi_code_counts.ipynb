```python
# gets ages of individuals' third column
def get_ages(row): # order is year, month, any number of columns representing dates
    month_dict = {'January':1,'February':2,'March':3,'April':4,'May':5,'June':6,'July':7,'August':8,'September':9,'October':10,'November':11,'December':12}
    birth_year = row[0]
    birth_month = month_dict[ row[1] ]
    
    if pd.isna(row[2]):
        return np.nan

    i = row[2].split("-")
    feat_year = int(i[0])
    feat_month = int(i[1])
    
    if feat_month >= birth_month:
        return feat_year - birth_year
    return feat_year - birth_year - 1
```

```python
# adds age_f.41270* columns to the dataframe that contains the ages codes were identified

byear_col = "f.34.0.0"
bmonth_col = "f.52.0.0"
icd10_col = "f.41270"
icd10_date_col = "f.41280"

start = datetime.now()
for feat_date in [c for c in df.columns.to_list() if icd10_date_col in c]:
    df["age_"+feat_date.replace(icd10_date_col, icd10_col)] = df[[byear_col, bmonth_col, feat_date]].apply(get_ages, axis=1)
print(datetime.now() - start)
```

```python
# given a dataframe count the number of individuals that have a certain code and age
# returns this count as well as a list of all the individuals with this code and age as well as at what columns this code occurs for them
# at age -1 age is not relevant in filtration

def get_count(dataframe, col_prefix, code, age = -1):
    indiv_w_code = []
    
    def indiv(row):
        cols = [c for c in dataframe.columns if col_prefix in c and "age_" not in c]
        temp = [row["IID"], row["FID"]]
        coded = False
        for i in cols:
            if age == -1:
                if not pd.isna(row[i]) and row[i] == code:
                    coded = True
                    temp.append([i, row["age_"+i]])
            else:
                if not pd.isna(row[i]) and row[i] == code and row["age_"+i] == age:
                    coded = True
                    temp.append([i, row["age_"+i]])
        if len(temp) > 2: indiv_w_code.append(temp)
    
    dataframe.apply(indiv, axis=1)
    
    return len(indiv_w_code), indiv_w_code
```

```python
byear_col = "f.34.0.0"
bmonth_col = "f.52.0.0"
icd10_col = "f.41270"
icd10_date_col = "f.41280"

start = datetime.now()
c, indiv_w_code = get_count(df, icd10_col, "H919")
print(datetime.now() - start)
```

```python
# get codes that exist in at least one individual more than once

keys = {}
def get_multiple(row):
    cols = [c for c in df.columns if icd10_col in c and "age_" not in c]
    counts = dict()
    for i in cols:
        if not pd.isna(row[i]) and row[i] not in counts.keys():
            counts[row[i]] = 1
        elif not pd.isna(row[i]):
            counts[row[i]] += 1
    #print(counts)
    max_key = max(counts, key = lambda x: counts[x])
    if counts[max_key] > 1: keys.append(max_key)
```

```python
df.apply(get_multiple, axis=1)
```
