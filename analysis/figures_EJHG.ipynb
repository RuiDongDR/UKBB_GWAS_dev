## H-aid

````r
# Mega-analysis
setwd('~/UKBiobank/results/REGENIE_results/results_burden_exome/101321_f3393_hearing_aid')
````

````r
f3393_m1_0.01 <- read.table(gzfile('090321_UKBB_Hearing_aid_f3393_expandedwhite_6436cases_96601ctrl_PC1_2_f3393.regenie_burden.Mask1.0.01.remove_sin.snp_stats.gz'), header=T, sep = "\t")
````

````r
head(f3393_m1_0.01)
````

````r
f3393_m2_0.01 <- read.table(gzfile('090321_UKBB_Hearing_aid_f3393_expandedwhite_6436cases_96601ctrl_PC1_2_f3393.regenie_burden.Mask2.0.01.remove_sin.snp_stats.gz'), header=T, sep = "\t")
````

````r
head(f3393_m2_0.01)
````

````r
joined_df <- merge(f3393_m1_0.01, f3393_m2_0.01, by = c("CHR","POS", "REF"), 
              all.x = TRUE, all.y = TRUE)
````

````r
head(joined_df, 100)
````

````r
df <- joined_df %>% separate(SNP.x, c("Gene1"), extra = "drop")
````

````r
df1 <- df %>% separate(SNP.y, c("Gen2"), extra = "drop")
````

````r
final_df <- df1 %>% 
    mutate(Gene = coalesce(Gene1,Gen2)) %>%
    select('Gene', 'CHR', 'POS', 'P.x', 'P.y')
````

````r
names(final_df)[names(final_df) == 'CHR'] <- 'Chromosome'
names(final_df)[names(final_df) == 'POS'] <- 'Position'
names(final_df)[names(final_df) == 'P.x'] <- 'Haid_LoF'
names(final_df)[names(final_df) == 'P.y'] <- 'Haid_LoF_missense'
````

````r
final_df
````

````r
SNPs <- list(
	final_df$Gene[final_df$Haid_LoF<1e-6],
	final_df$Gene[final_df$Haid_LoF_missense<1e-6]
)
CMplot(final_df, plot.type="m",multracks=TRUE,col=c("darkblue", "darkorange"), threshold=c(2.5e-6),threshold.lty=c(2), 
        threshold.lwd=c(1), threshold.col=c("red"), amplify=TRUE,bin.size=1e6,
        signal.col=c("darkblue","darkorange"),ylim=c(0,21),
        signal.cex=1, file="jpg",memo="",dpi=300,file.output=TRUE,verbose=TRUE,
        highlight=SNPs, highlight.text=SNPs, highlight.text.cex=1.4)
#Note: if you are not supposed to change the color of signal, 
#          please set signal.col=NULL and highlight.col=NULL.
````
