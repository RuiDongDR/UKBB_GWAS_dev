# Manhattan plot for RAP results

This workflow is designed to annotate the manhattan plots generated from results derived from RAP - UKB processing

```sos
R.version
```

```sos
library(topr)
library(dplyr)
```

## Haid univariate

```sos
df <- read.delim('~/UKBiobank/RAP/results/autosomal/univariate/Haid/f3393.regenie', sep = ' ')
```

```sos
df[,"LOG10P"] <-as.numeric(unlist(df[,"LOG10P"]))
df$P <- 10^-df$LOG10P
df_sig <- df %>% filter(P<= 1e-03)
```

```sos
df_gws<- df %>% filter(P<= 5e-08)
```

```sos
df_gws$ID_2 <- paste(df_gws$CHROM, df_gws$GENPOS, sep = "_")
```

```sos
df_gws
```

```sos
df_new <- df_sig %>% rename (POS = GENPOS) %>%
    select(CHROM,POS,P)
```

```sos
manhattan(df_new, annotate=5e-8)
```

```sos
get_lead_snps(df_new)
```

```sos
lead_snps_haid = get_lead_snps(df_new) %>% annotate_with_nearest_gene()
```

```sos
lead_snps_haid
```

```sos
write.csv(lead_snps_haid, file = "~/UKBiobank/RAP/results/autosomal/univariate/Haid/Haid_lead_snps_topr.csv", row.names = FALSE, quote=FALSE)
```

```sos
regionplot(df_new, gene="SLC22A7",annotate=5e-09,region_size = 1000000)
```

```sos
regionplot(df_new, gene="SYNJ2", annotate=5e-09)
```

```sos
regionplot(df_new, gene="FILIP1", annotate=5e-09)
```

```sos
regionplot(df_new, gene="BAIAP2L2")
```

```sos
regionplot(df_new, gene="TRIOBP")
```

```sos
regionplot(df_new, gene="LINC02210-CRHR1", annotate=5e-09)
```
